### 설문조사

1. 볼 색이 얼굴 전체를 대표하는 색으로 지정할 수 있는가?


- 얼굴 전체를 대표하는 색이 볼 색이 맞는가에 대해 귀무가설, 대립가설을 세운다.


- 연예인 사진의 화이트 밸런스를 조절하여 볼과 턱의 HSV 값을 뽑아낸다.


      설문지 상에는 얼굴 전체 사진, 볼 색, 턱 색을 나열해 두고 볼과 턱 중 어느 색이 얼굴 전체 색과 더 가까운지에 대해 조사한다.
  
  
2. 퍼스널 컬러는 의미가 있는 것인가?


- 논문에서 언급한 퍼스널 컬러의 기준이 의미가 있는것인지에 대해 귀무가설, 대립가설을 세운다.


- 퍼스널 컬러 타입 기준별로 보색, 삼각 보색, 이중 보색을 알아보고 피부색을 가장 밝게 보여주는 색상은 무엇인지 조사한다. 
  

- 퍼스널 컬러 타입 기준에 반대되는 색상들을 알아보고 이 색상들이 피부색에 어떤 효과를 나타내는지 조사한다.
  
  
- 퍼스널 컬러 타입 기준의 경계선에 있는 색상들을 알아보고 이 색상들이 피부색에 어떤 효과를 나타내는지 조사한다.


#### 설문조사는 Naver Form 이나 프린트하여 배포할 예정
