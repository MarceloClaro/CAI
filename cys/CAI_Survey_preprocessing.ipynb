```python
## 밝은색 설문 결과 불러오기
AS_hueChanged_bright = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문1.xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="C, D, M, N, O, P, Q, R")

AS_hueChanged_bright.rename(columns={AS_hueChanged_bright.columns[0]:"sex",
                                AS_hueChanged_bright.columns[1]:"age_class",
                                AS_hueChanged_bright.columns[2]:"comp",
                                AS_hueChanged_bright.columns[3]:"triangle1",
                                AS_hueChanged_bright.columns[4]:"triangle2",
                                AS_hueChanged_bright.columns[5]:"doubleness1",
                                AS_hueChanged_bright.columns[6]:"doubleness2",
                                AS_hueChanged_bright.columns[7]:"doubleness3"}, inplace=True)

columns = ["comp", "triangle1", "triangle2", "doubleness1", "doubleness2", "doubleness3"]


## 값 변경 반복문
# sex, age = [], []
for i in range(int(AS_hueChanged_bright.shape[0])):
    ## sex : Male, Female 리스트 담기
    if AS_hueChanged_bright["sex"][i] == "남성":
        AS_hueChanged_bright["sex"][i] = "Male"
    else:
        AS_hueChanged_bright["sex"][i] = "Female"
        
    ## age_class : 10, 20, 30, 40, 50으로 리스트 담기
    if AS_hueChanged_bright["age_class"][i] == "10대":
        AS_hueChanged_bright["age_class"][i] = "10"
    elif AS_hueChanged_bright["age_class"][i] == "20대":
        AS_hueChanged_bright["age_class"][i] = "20"
    elif AS_hueChanged_bright["age_class"][i] == "30대":
        AS_hueChanged_bright["age_class"][i] = "30"
    elif AS_hueChanged_bright["age_class"][i] == "40대":
        AS_hueChanged_bright["age_class"][i] = "40"
    else:
        AS_hueChanged_bright["age_class"][i] = "50"
        
        
## 선택항목 별 부위 매칭 (①-색+72º, ②-색+144º, ③-original, ④-색+216º, ⑤-색+288º)
for i in range(int(AS_hueChanged_bright.shape[0])):
    for col in columns:
        if AS_hueChanged_bright[col][i] == "①":
            AS_hueChanged_bright[col][i] = "72"
        elif AS_hueChanged_bright[col][i] == "②":
            AS_hueChanged_bright[col][i] = "144"
        elif AS_hueChanged_bright[col][i] == "③":
            AS_hueChanged_bright[col][i] = "original"
        elif AS_hueChanged_bright[col][i] == "④":
            AS_hueChanged_bright[col][i] = "216"
        else:
            AS_hueChanged_bright[col][i] = "288"
```

### 연령, 성별에 따른 색 고정, 명도/채도 변경 팔레트 설문결과 전처리

```python
## 밝은색 설문 결과 불러오기
AS_hueFixed_bright = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문1.xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="C, D, S, T, U, V, W, X")

AS_hueFixed_bright.rename(columns={AS_hueFixed_bright.columns[0]:"sex",
                                AS_hueFixed_bright.columns[1]:"age_class",
                                AS_hueFixed_bright.columns[2]:"comp",
                                AS_hueFixed_bright.columns[3]:"triangle1",
                                AS_hueFixed_bright.columns[4]:"triangle2",
                                AS_hueFixed_bright.columns[5]:"doubleness1",
                                AS_hueFixed_bright.columns[6]:"doubleness2",
                                AS_hueFixed_bright.columns[7]:"doubleness3"}, inplace=True)

columns = ["comp", "triangle1", "triangle2", "doubleness1", "doubleness2", "doubleness3"]


## 값 변경 반복문
# sex, age = [], []
for i in range(int(AS_hueFixed_bright.shape[0])):
    ## sex : Male, Female 리스트 담기
    if AS_hueFixed_bright["sex"][i] == "남성":
        AS_hueFixed_bright["sex"][i] = "Male"
    else:
        AS_hueFixed_bright["sex"][i] = "Female"
        
    ## age_class : 10, 20, 30, 40, 50으로 리스트 담기
    if AS_hueFixed_bright["age_class"][i] == "10대":
        AS_hueFixed_bright["age_class"][i] = "10"
    elif AS_hueFixed_bright["age_class"][i] == "20대":
        AS_hueFixed_bright["age_class"][i] = "20"
    elif AS_hueFixed_bright["age_class"][i] == "30대":
        AS_hueFixed_bright["age_class"][i] = "30"
    elif AS_hueFixed_bright["age_class"][i] == "40대":
        AS_hueFixed_bright["age_class"][i] = "40"
    else:
        AS_hueFixed_bright["age_class"][i] = "50"
        
        
## 선택항목 별 부위 매칭 (①-채도(-5)/명도(-5), ②-채도(-5)/명도(+5), ③-original, ④-채도(+5)/명도(+5), ⑤-채도(+5)/명도(-5))
for i in range(int(AS_hueFixed_bright.shape[0])):
    for col in columns:
        if AS_hueFixed_bright[col][i] == "①":
            AS_hueFixed_bright[col][i] = "-S-V"
        elif AS_hueFixed_bright[col][i] == "②":
            AS_hueFixed_bright[col][i] = "-S+V"
        elif AS_hueFixed_bright[col][i] == "③":
            AS_hueFixed_bright[col][i] = "original"
        elif AS_hueFixed_bright[col][i] == "④":
            AS_hueFixed_bright[col][i] = "+S+V"
        else:
            AS_hueFixed_bright[col][i] = "+S-V"    
```

### 기본항목 전처리

```python
# 기본항목 설문결과 (모집단 분리하지 않은 항목) 데이터 준비

## 밝은색 설문 결과 불러오기
elementaryItem1 = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문1.xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="C, D, E, F, G, H")
elementaryItem1.rename(columns={elementaryItem1.columns[0]:"sex",
                          elementaryItem1.columns[1]:"age_class",
                          elementaryItem1.columns[2]:"definition",
                          elementaryItem1.columns[3]:"need",
                          elementaryItem1.columns[4]:"purpose",
                          elementaryItem1.columns[5]:"interest"}, inplace=True)

## 조화색 설문 결과 불러오기
elementaryItem2 = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문2.xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="C, D, E, F, G, H")
elementaryItem2.rename(columns={elementaryItem2.columns[0]:"sex",
                          elementaryItem2.columns[1]:"age_class",
                          elementaryItem2.columns[2]:"definition",
                          elementaryItem2.columns[3]:"need",
                          elementaryItem2.columns[4]:"purpose",
                          elementaryItem2.columns[5]:"interest"}, inplace=True)


## 두 결과 합치기
elementaryItem = pd.concat([elementaryItem1, elementaryItem2], join='outer')


## 인덱스 조정
elementaryItem.reset_index(inplace=True)
del elementaryItem['index']


## 값 변경 반복문
# sex, age, definition, need, purpose_colthes, purpose_cosmetic, purpose_etc = [], [], [], [], [], [], []
for i in range(int(elementaryItem.shape[0])):
    ## sex : Male, Female 리스트 담기
    if elementaryItem["sex"][i] == "남성":
        elementaryItem["sex"][i] = "Male"
    else:
        elementaryItem["sex"][i] = "Female"
        
    ## age_class : 10, 20, 30, 40, 50으로 리스트 담기
    if elementaryItem["age_class"][i] == "10대":
        elementaryItem["age_class"][i] = "10"
    elif elementaryItem["age_class"][i] == "20대":
        elementaryItem["age_class"][i] = "20"
    elif elementaryItem["age_class"][i] == "30대":
        elementaryItem["age_class"][i] = "30"
    elif elementaryItem["age_class"][i] == "40대":
        elementaryItem["age_class"][i] = "40"
    else:
        elementaryItem["age_class"][i] = "50"
    
    ## definition : bright, harmony로 리스트 담기
    if ("생기" in elementaryItem["definition"][i]):
        elementaryItem["definition"][i] = "bright"
    else:
        elementaryItem["definition"][i] = "harmony"
    
    ## need :  bright, harmony로 리스트 담기
    if ("생기" in elementaryItem["need"][i]):
        elementaryItem["need"][i] = "bright"
    else:
        elementaryItem["need"][i] = "harmony"
    
    ## purpose : clothes, cosmetic, etc로 리스트 담기
#     if elementaryItem["purpose"][i] == "옷을 구매하는 경우에 활용":
#         elementaryItem["purpose"][i] = "clothes"
#     elif elementaryItem["purpose"][i] == "화장하는 경우에 활용":
#         elementaryItem["purpose"][i] = "cosmetic"
#     else:
#         elementaryItem["purpose"][i] = "etc"

    if ("옷" in elementaryItem["purpose"][i]) & ("화장" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "clothes&cosmetic"
    elif ("옷" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "clothes"
    elif ("화장" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "cosmetic"
    else:
        elementaryItem["purpose"][i] = "etc"
```

```python
# 기본항목 전처리 결과 저장
elementaryItem.to_csv("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/element.csv")
```
