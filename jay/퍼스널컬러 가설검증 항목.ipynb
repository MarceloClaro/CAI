### 퍼스널컬러 가설검증 항목

1. 퍼스널 컬러를 분류하는 안면부위
    - 논문내용
       > * 명도 : 모든 부위에서 화장 전후 명도 차이가 존재함   
       > * 적색도 : 얼굴 전체 부위에서 화장 전후의 적색도 차이가 존재하지 않음   
       > * 황색도 : 이마. 턱에서 색수치 차이가 존재하고, 볼에서는 존재하지 않음   
       > → 여성들이 이마와 턱에 비해 볼의 황색도를 보정하지 않는다는 분석결과를 통하여 여성이 볼의 스킨 톤을 자신의 전체 스킨 톤을 대표한다고 생각하고, 볼의 스킨톤을 살리는 것이 개성을 나타내는 가장 자연스러운 방법
    - 의문점
       - 적색도 : 차이 셋다 별로 없음
       - 황색도 : 이마, 볼, 턱 순서로   
          평균      0.24 / 0.2 / 0.55   
          표준편차 -0.23 / 0.14 / 0.42 (하단 표 참고)   
       → 이마와 볼의 화장 전후 황색도 차이가 0.1, 의미있는 차이라고 할 수 있는가   
       → 사람을 더 밝아보이고 생기있어보이게 하는 목적의 퍼스널컬러 이론에서 '여성이 자신의 전체 스킨 톤을 대표하는 부위를 볼로 생각한다'를 스킨 톤 측색의 안면부위 근거로 삼을 수 있는가   
       → 타인이 인식할때 가장 대표되는 스킨 톤을 퍼스널컬러 측색을 위한 안면부위로 특정해야하는 것이 아닌가   
        1. 가설검증
            - 귀무가설 : 볼이 얼굴을 대표하는 스킨 톤이 맞는가
            - 대립가설 : 볼이 아닌 다른 부위(턱, 이마, 눈, 얼굴전체평균)가 얼굴을 대표하는 스킨 톤이 맞는가
            - 검증방법
                - 샘플 사진의 화이트 밸런스를 조절, 이마, 볼, 눈, 턱, 얼굴 전체의 각  평균 색 값 추출
                - 사진과 각 색을 팔레트로 비교하여 어느 색이 가장 샘플사진에 가까운 색인지 설문
                - 설문 결과를 가지고 검증   
   
   
2. 퍼스널 컬러는 의미가 있는 것인가?
    - 의문점
        - 퍼스널컬러는 색채이론의 근거하여 사람을 밝아보이게 하는데 목적을 둔다
        - 색채이론 상 색을 밝아보이게 하는 것은 색대비, 명도대비, 채도대비 이론에 의함
        - 기준색을 샘플로, 이론에 따른 색을 비교했을 때 사람들은 실제로 기준색을 밝아보인다고 느끼는가
        - 기준색을 샘플로, 이론과 다른 의미없는 색을 비교했을 때 밝아보인다고 느끼는가
        - 스킨 톤(피부색)은 환경에 따라 달라지는 색이기도 하지만, 디지털이나 인쇄물과 달리 늘 바뀌는 색이기도 하며, 퍼스널컬러 이론의 목적이 타인에게 어떻게 보이는가에 있기 때문에, 타인의 색 인식에 차이가 얼마나 있는가도 퍼스널컬러 이론이 의미가 있는가에 영향을 끼치는 요인이라고 볼 수 있지 않은가
    
        1. 가설검증
            - 귀무가설 : 기준색에 대하여 대비이론에 따르는 색(H고정)을 가장 밝아보인다고 느낀다
            - 대립가설 : 기준색에 대하여 대비이론에 따르는 색(H고정)을 가장 밝아보인다고 느끼지 않는다
            - 검증방법
                - 기준색을 샘플로 해 보색, 삼각 보색, 이중 보색을 추출
                - 각 보색들의 색 값을 고정하고 명도, 채도를 각기 달리한 색을 비교
                - 각 보색과 조정한 색들 중 가장 밝아보이는 색을 조사
        2. 가설검증
            - 귀무가설 : 기준색에 대하여 대비이론에 따르는 색(S,V고정)을 가장 밝아보인다고 느낀다
            - 대립가설 : 기준색에 대하여 대비이론에 따르는 색(S,V고정)을 가장 밝아보인다고 느끼지 않는다
            - 검증방법
                - 기준색을 샘플로 해 보색, 삼각 보색, 이중 보색을 추출
                - 각 보색들의 명도, 채도 값을 고정하고 색값을 각기 달리한 색을 비교
                - 각 보색과 조정한 색들 중 가장 밝아보이는 색을 조사
        3. 가설검증
            - 귀무가설 : 기준색을 밝아보이게 하는 색과 기준색과 조화롭다고 느끼는 색이 일치한다
            - 대립가설 : 기준색을 밝아보이게 하는 색과 기준색과 조화롭다고 느끼는 색이 일치하지 않는다
            - 검증방법
                - 위 두가지 가설검증 항목을 밝아보이는게 아닌 조화로운 색을 고르는 것으로 모집단이 겹치지 않게 별도조사
        4. 가설검증
            - 귀무가설 : 색을 인식하는 사람의 연령, 성별에 따라 색을 인식하는 차이가 있다
            - 대립가설 : 색을 인식하는 사람의 연령, 성별에 따라 색을 인식하는 차이가 없다
            - 검증방법
                - 설문조사 항목에 연령과 성별 항목을 추가
                - 각 설문 문항들에 대하여 색 인식에 대한 차이가 존재하는지 조사
   
   
3. 퍼스널컬러에 대한 인식
    - 의문점
        - 사람들이 인식하는 퍼스널컬러는 피부색을 밝아보이게 만드는 색인가, 피부색과 어울리는 색인가
        - 사람들은 퍼스널컬러와 무관하게 옷, 화장품 색 등에 의해서 피부색이 밝아보이는 것을 선호하는가   
          아니면 피부색과 더 조화로운/어울리는 색의 옷, 화장품의 색을 선호하는가
        
        1. 가설검증
            - 귀무가설 : 사람들이 인식하는 퍼스널 컬러는 피부색을 밝아보이게 만드는 색이다
            - 대립가설 : 사람들이 인식하는 퍼스널 컬러는 피부색과 어울리는 색이다
            - 검증방법
                - 설문조사 항목에 퍼스널컬러에 대한 인식 여부 항목 추가

        2. 가설검증
            - 귀무가설 : 사람들이 옷, 화장품의 색을 통해 피부색이 밝아보이는 것을 선호한다
            - 대립가설 : 사람들이 옷, 화장품의 색이 피부색과 조화로운 것을 선호한다
